# Generating Yelp URL

```python
def inputParams():
    desc = input('description: ').replace(' ','%20')
    loc = input('location: ').replace(' ','%20')
    params = {'description':desc, 'location':loc}
    return params

def genURL(params):
    URL = 'https://www.yelp.com/search?'
    find_description = 'find_desc='+str(params['description'])
    find_location = '&find_loc='+str(params['location'])
    URL+=find_description+find_location
    return URL
```

```python
params = inputParams()
```

```python
gen_url = str(genURL(params))
print(gen_url)
```

# Generate each restaurant across a given page

```python
import bs4 as bs
import urllib.request as url
```

```python
source = url.urlopen(gen_url)
```

```python
page_soup = bs.BeautifulSoup(source, 'html.parser')
```

```python
mains = page_soup.find_all("a", {"class":"link__09f24__1kwXV link-color--inherit__09f24__3PYlA link-size--inherit__09f24__2Uj95"})
```

```python
links = []
for link in mains:
    links.append(link['href'])

links

# https://www.yelp.com/biz/grilld-harbourside-sydney?osq=vegan+burgers
```
